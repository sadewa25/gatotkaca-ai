## Import Libraries

```python
import pandas as pd
```

```python
# URL of the CSV file from Open-Meteo
url = "open-meteo-sby.csv"

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(url, delimiter=",")

# Display the DataFrame
df.keys()
```

```python
# sorting date
df_sorted = df.sort_values(by='time')
df_sorted.head()
```

```python
df_sorted.tail()
```

```python
# check if there are nullable values
df_sorted.isna().sum()
```

```python
# Remove rows where 'weather_code' is NaN
df_cleaned = df_sorted.dropna(subset=['weather_code (wmo code)'])
df_cleaned.isna().sum()
```

```python
# replace is na in duration with -1
df_cleaned['sunshine_duration (s)'] = df_cleaned['sunshine_duration (s)'].fillna(-1)
df_cleaned.isna().sum()
```

```python
# how to rename column weather_code (wmo code) with wmo_code
df_cleaned.rename(columns={'weather_code (wmo code)': 'wmo_code'}, inplace=True)
df_cleaned.rename(columns={'temperature_2m_max (°C)': 'temp_2m_max'}, inplace=True)
df_cleaned.rename(columns={'temperature_2m_min (°C)': 'temp_2m_min'}, inplace=True)
df_cleaned.rename(columns={'sunrise (iso8601)': 'sunrise'}, inplace=True)
df_cleaned.rename(columns={'sunset (iso8601)': 'sunset'}, inplace=True)
df_cleaned.rename(columns={'daylight_duration (s)': 'daylight'}, inplace=True)
df_cleaned.rename(columns={'sunshine_duration (s)': 'sunshine'}, inplace=True)
```

```python
df_cleaned.head()
```

```python
df_cleaned.tail()
```

```python
# reset the index
df_cleaned.reset_index(drop=True, inplace=True)
```

```python
# i want to export the dataframe to csv
df_cleaned.to_csv('open-meteo-sby-normalized.csv', index=False)
```

```python
# df_sql from df_cleaned
df_sql = df_cleaned.copy()
```

```python
# convert string date to date
# Ensure the date column is of type datetime64
df_sql['time'] = pd.to_datetime(df_sql['time'], errors='coerce')
```

```python
df_sql.head(), df_cleaned.isna().sum()
```

```python
# convert string to datetime
# Step 2: Use pd.to_datetime to convert the column to datetime
df_sql['sunrise'] = pd.to_datetime(df_sql['sunrise'], format='%Y-%m-%dT%H:%M', errors='coerce')
df_sql['sunset'] = pd.to_datetime(df_sql['sunset'], format='%Y-%m-%dT%H:%M', errors='coerce')
```
